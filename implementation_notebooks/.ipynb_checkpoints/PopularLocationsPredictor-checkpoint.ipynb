```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Load the dataset (e.g., TripAdvisor or Yelp dataset)
# Replace 'dataset.csv' with your dataset file
data = pd.read_json(r"..\datasets\yelp_academic_dataset_business.json",lines=True)
```

```python


# Step 1: Filter data for the user-specified location
user_location = "Paris"  # Example location
user_lat = 48.8566  # Example latitude for Paris
user_lon = 2.3522   # Example longitude for Paris

# Filter data for the specified city (case-insensitive)
filtered_data = data[data['city'].str.lower() == user_location.lower()]

# Check if filtered_data is empty
if filtered_data.empty:
    print(f"No data found for location: {user_location}. Please try another location.")
else:
    # Step 2: Feature Engineering
    # Create features like popularity, average rating, and distance
    filtered_data['popularity'] = filtered_data['review_count']  # Number of reviews as popularity
    filtered_data['average_rating'] = filtered_data['stars']  # Average rating (stars)
    filtered_data['distance'] = np.sqrt(
        (filtered_data['latitude'] - user_lat)**2 + (filtered_data['longitude'] - user_lon)**2
    )  # Distance from user-specified location

    # Step 3: Prepare data for the regression model
    X = filtered_data[['popularity', 'average_rating', 'distance']]
    y = filtered_data['stars']  # Target variable (average rating)

    # Normalize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    # Step 4: Build the Regression Model (Linear Regression)
    model = LinearRegression()
    model.fit(X_train, y_train)  # Train the model

    # Step 5: Make predictions
    y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Mean Absolute Error (MAE): {mae}")

    # Step 6: Rank locations, attractions, and hotels
    filtered_data['predicted_score'] = model.predict(X_scaled)
    top_locations = filtered_data.sort_values(by='predicted_score', ascending=False).head(10)

    # Step 7: Display results
    print("\nTop Popular Locations:")
    print(top_locations[['name', 'popularity', 'average_rating', 'distance']])

    # Filter and display top attractions (assuming 'categories' contains 'attraction')
    print("\nTop Best-Visited Places:")
    top_attractions = filtered_data[filtered_data['categories'].str.contains('attraction', case=False, na=False)].sort_values(by='predicted_score', ascending=False).head(10)
    print(top_attractions[['name', 'popularity', 'average_rating', 'distance']])

    # Filter and display top hotels (assuming 'categories' contains 'hotel')
    print("\nTop Best Hotels:")
    top_hotels = filtered_data[filtered_data['categories'].str.contains('hotel', case=False, na=False)].sort_values(by='predicted_score', ascending=False).head(10)
    print(top_hotels[['name', 'popularity', 'average_rating', 'distance']])
```
